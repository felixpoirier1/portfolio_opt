```python
def téléchargerDf(PATH):
    # génère une liste avec le nom de chaque action
    stock_list = os.listdir(PATH)

    # nous enlevons l'extension .csv du nom des valeurs de "stock_list"
    stock_list = [stock_list[i].replace('.csv','') for i in range(len(stock_list))][:30]


    # initialisation d'un dictionnaire où seront stockés les DataFrames
    dict_stocks = {i: None for i in stock_list}

    # création des DataFrame
    for i in stock_list:
        # un fichier s'appelle ".DS_Store", il faut l'ignorer
        if i != '.DS_Store':
            # création des DataFrame
            dict_stocks[i] = pd.read_csv(f'{PATH}{i}.csv')

        # pour garder un échantillon élevé nous conservons seulement
        # les DataFrames qui ont 1260 jours soit depuis février 2017 (5 ans)
        try:
            if len(dict_stocks[i]) != 1260:
                del dict_stocks[i]
        except:
            pass
        
    '''pour supprimer toutes les valeurs 'None' : 
    (des valeurs de "dict_stocks" n'ayant pas
    été capable de générer un DataFrame)'''
    for i in dict_stocks:
        if type(dict_stocks[i]) == None:
            del dict_stocks[i]
    
    return dict_stocks
    
```

```python
def créerRendement(dict_stocks):
    # fonction qui retourne un data frame rendement
    def rendement(df):
        df['daily_return'] = (df['Close'] / df['Close'].shift(1))-1
        return df  
    
    # fonction qui retourne un data frame du rendement cumulatif
    def rendementCumul(df):
        df['cum_return'] = (1+df['daily_return']).cumprod()
        return df
    
    for i in dict_stocks:
        dict_stocks[i] = rendement(dict_stocks[i])
        dict_stocks[i] = rendementCumul(dict_stocks[i])
    return dict_stocks
```

# Matrice des covariances

Voici tous les opérations que nous devrons reproduire:
<img src='IMAGES/variance.png' width="400" height="400">

                        source: Cours de programmation de modèles quantitatifs en finance

```python
def créerMatCovar(dict_stocks):
    # première opérations consiste à soustraire de chaque rendement la moyenne des rendements
    df_var = pd.DataFrame()

    for i in dict_stocks:
        # soustractions avec conditions try pour éviter les érreurs
        try:
            df_var[i] = dict_stocks[i].loc[:]['daily_return'] - dict_stocks[i].loc[:]['daily_return'].mean()
        except:
            pass
        
    # le rendement de la première journée ne peux être calculé donc ces valeurs sont ignorées
    df_var = df_var.loc[1:]

    # utilisation des arrays de numpy qui permettent de faire des calculs matriciels
    mat_var = df_var.to_numpy()

    mat_covar = 1/len(mat_var)*np.matmul(np.transpose(mat_var),mat_var)
    
    return mat_covar

```

# Optimisation

Voici le modèle d'optimisation qui nous permettera de minimiser la variance du portfolio:
<img src='IMAGES/optimisation.png' width="200" height="400">


                     
<img src='IMAGES/Optimisation_matrice.png' width="800" height="400">

                    source: cours de programmation de modèles quantitatifs en finance

______________
__Condition 1: le produit matriciel entre les poids et les rendements, doit être égaulau rendement désiré__

__Condition 2: le produit matriciel entre e et les poids doivent être égal à 1, donc tout l'argent dans notre portfolio doit être utilisé__

__En somme, nous devons trouver un scalaire x qui minimise la variance de notre portefeuille__
______________
_x: Vecteur des poids de chaque actif_

_S: Matrice des covariances "mat_covar"_

_r: Vecteur des rendements de chaque actif_

_rpf: Scalaire représentant le rendement désiré_

_e: Vecteur [1, 1, 1, ..., 1]_

```python
def vecR(dict_stocks):
    r = []
    for i in dict_stocks:
        # soustractions avec conditions try pour éviter les erreurs
        try:
            # itération dans les DataFrames pour trouver le rendement moyen de chaque action
            # puis on le rajoute à la liste r (ne pas utiliser les index 0 puisque == NaN)
            r.append(dict_stocks[i].loc[1:]['daily_return'].mean())
        except:
            pass

    # transformation de la liste r en numpy array (vecteur)
    vec_r = np.array([r])
    
    return vec_r

# liste des rendements
def créerMatCovarAugmentée(mat_covar, vec_r, dict_stocks):

    # création d'une liste contenant n fois 1 (n étant le nombre d'actions)
    e = [1 for i in range(len(mat_covar))]

    # transformation de la liste e en numpy array (vecteur)
    vec_e = np.array(e)

    # concatenation entre la matrice des covariances et les vecteurs r et e transposé
    # axis = 1 dit à numpy de de rajouter "r" et "e" en tant que colonnes (à la fin)
    mat_covar_aug = np.concatenate((mat_covar, np.transpose(vec_r)), axis = 1)
    mat_covar_aug = np.concatenate((mat_covar_aug, np.transpose([vec_e])), axis = 1)


    # ceci permet de savoir combien de 0 il faut rajouter aux deux vecteurs r et e
    # comme pour le côté inférieur droit de la matrice des covariances augmentés (voir image en haut)
    zero_to_append = len(mat_covar_aug[0])-len(dict_stocks)

    vec_r_aug = vec_r
    vec_e_aug = vec_e

    # itération de "zero_to_append" et rajout des "0" aux deux vecteurs
    for i in range(zero_to_append):
        vec_r_aug = np.append(vec_r_aug,0)
        vec_e_aug = np.append(vec_e_aug,0)

    # concatenation entre la matrice des covariances et les vecteurs r et e
    # axis = 0 dit à numpy de de rajouter "r" et "e" en tant que lignes (à la fin)
    mat_covar_aug = np.concatenate((mat_covar_aug, [vec_r_aug]), axis = 0)
    mat_covar_aug = np.concatenate((mat_covar_aug, [vec_e_aug]), axis = 0)
    
    return mat_covar_aug
```

# Ici il faudra inscrire le rendement désiré

```python
# fonction qui prends une rendement spécifique
# et qui le transforme en rendement continu journalier
def convRendement(rend):
    
    # rendement_continu = ln(1+r/période)
    
    if rend[-1] == "a":
        rend = math.log(1+float(rend[:-1])/365)
    elif rend[-1] == "m":
        rend = math.log(1+float(rend[:-1])/30)
    elif rend[-1] == "j":
        rend = math.log(1+float(rend[:-1]))
        
    # pour retourner une valeur prédéfinie
    elif rend == "p":
        rend = math.log(1+0.001)
    
    # si le rendement est inadéquat retourne une erreur
    else:
        raise TypeError('Vous devez rajouter un "a", "m" ou "j" à la fin du rendement')

    return rend

def créerVecCond(mat_covar_aug, rendement):
    rendement_continu = convRendement(rendement)

    # création du vecteur des valeurs des conditions de la même taille de les vecteurs r et e
    cond = [0 for i in range(len(mat_covar_aug)-2)]

    # rajout des valeurs des deux contraintes "rpf" et 1
    cond.extend([rendement_continu, 1])

    # transformation en numpy.array
    vec_cond = np.array([cond])

    return vec_cond
```

```python
def créerVecPoids(mat_covar_aug, vec_cond):
    # calcul de l'inverse de la matrice des covariances augmentée
    mat_covar_aug_inv = np.linalg.inv(mat_covar_aug)

    # multiplication entre l'inverse de la matrice des covariances augmentée
    # et le vecteur des conditions (ceci retourne le vecteur recherché)
    vec_rech = np.matmul(mat_covar_aug_inv,np.transpose(vec_cond))

    # le vecteur recherché contient des valeurs pour chaque condition 
    # (celle-ci ne sont pas importantes) nous voulons seulement les poids
    vec_poids = np.transpose(vec_rech[:-2])

    return vec_poids
```

# Formules pour trouver les statistiques du portfolio

<img src='IMAGES/Formules.png' width="800" height="400">

```python
#fonction qui retourne un tuple (rendement, écart-type)
def rendVol(vec_r, vec_poids, mat_covar):
    # rendement du portfolio trouvé en multipliant le vecteur des poids au vecteur des rendements
    rendement_portfolio = np.matmul(vec_r, np.transpose(vec_poids))[0][0]

    # variance du portfolio trouvé en multipliant le vecteur des poids
    # à la matrice des covariances puis encore par le vecteur des poids
    variance_portfolio = np.matmul(np.matmul(vec_poids, mat_covar), np.transpose(vec_poids))[0][0]

    # écart-type est la racine de la variance
    écart_type_portfolio = math.sqrt(variance_portfolio)

    return (rendement_portfolio, écart_type_portfolio)
```

```python
def produireStats(direction, rendement):
    dict_direction = {'1': "ALIMENTAIRE",
                  '2': "COMMODITÉS",
                  '3': "CONSOMMATION", 
                  '4': "ÉNERGIE",
                  '5': "FINANCE",
                  '6': "TECHNOLOGIE"}

    PATH = f"STOCKS/{dict_direction[direction]}/"
    
    dict_stocks = téléchargerDf(PATH)
    
    dict_stocks = créerRendement(dict_stocks)
    mat_covar = créerMatCovar(dict_stocks)

    vec_r = vecR(dict_stocks)

    mat_covar_aug = créerMatCovarAugmentée(mat_covar, vec_r, dict_stocks)
    
    vec_cond = créerVecCond(mat_covar_aug, rendement)
    
    vec_poids = créerVecPoids(mat_covar_aug, vec_cond)
    
    # rendement du portfolio trouvé en multipliant le vecteur des poids au vecteur des rendements
    rendement_portfolio = np.matmul(vec_r, np.transpose(vec_poids))[0][0]

    # variance du portfolio trouvé en multipliant le vecteur des poids
    # à la matrice des covariances puis encore par le vecteur des poids
    variance_portfolio = np.matmul(np.matmul(vec_poids, mat_covar), np.transpose(vec_poids))[0][0]

    # écart-type est la racine de la variance
    écart_type_portfolio = math.sqrt(variance_portfolio)

    return (rendement_portfolio, écart_type_portfolio)
```

```python
def produirePoids(direction, rendement):
    dict_direction = {'1': "ALIMENTAIRE",
                  '2': "COMMODITÉS",
                  '3': "CONSOMMATION", 
                  '4': "ÉNERGIE",
                  '5': "FINANCE",
                  '6': "TECHNOLOGIE"}
    
    PATH = f"STOCKS/{dict_direction[direction]}/"
    
    dict_stocks = téléchargerDf(PATH)
    
    dict_stocks = créerRendement(dict_stocks)
    
    mat_covar = créerMatCovar(dict_stocks)

    vec_r = vecR(dict_stocks)

    mat_covar_aug = créerMatCovarAugmentée(mat_covar, vec_r, dict_stocks)
    
    vec_cond = créerVecCond(mat_covar_aug, rendement)
    
    vec_poids = créerVecPoids(mat_covar_aug, vec_cond)
    
    liste_poids = [i for i in vec_poids[0]]

    # DataFrame des poids
    liste_stocks = [i for i in dict_stocks]
    df_poids = pd.DataFrame(liste_poids, columns = ["pct"], index=liste_stocks).transpose()
    
    return df_poids

```

```python
# Créer un dictionnaire composé des DF des titres associés au secteur choisi par l'utilisateur à l'aide des fonctions
# 'téléchargerDf' et 'créerRendement'
secteur_dict = {'1': "ALIMENTAIRE",'2': "COMMODITÉS",'3': "CONSOMMATION", '4': "ÉNERGIE",'5': "FINANCE",'6': "TECHNOLOGIE"}

stocks_dict = téléchargerDf("STOCKS/{0}/".format(secteur_dict[direction]))

diction = créerRendement(stocks_dict)

print(diction)
```

```python
# Calculer le retour annuel individuel de chaque titre du portefeuille créé pour l'utilisateur

# Dictionnaire contenant les différents secteurs disponibles pour la création de portefeuille
secteur_dict = {'1': "ALIMENTAIRE",'2': "COMMODITÉS",'3': "CONSOMMATION", '4': "ÉNERGIE",'5': "FINANCE",'6': "TECHNOLOGIE"}

# Créer un premier dictionnaire avec les titres comme clés (ex: ADM) et leurs DF correspondants comme valeurs
# Ce dictionnaire changera en foction du secteur ('direction') choisi par l'utilisateur
cpny_dict = téléchargerDf("STOCKS/{0}/".format(secteur_dict[direction]))

# Ajouter les colonnes de rendement journalier et cumulatif aux DF qui sont les valeurs du dictionnaire 'cpny_dict'
full_dict = créerRendement(cpny_dict)

# Créer un DF: Le titre de chaque colonne sera un actif du portefeuille; Les valeurs de la colonne = 'daily_return' 
daily_df = pd.DataFrame()

# Compteur qui sera utilisé dans la boucle qui suit
i = 0

# Utiliser une boucle 'for' pour itérer sur les titres (clés) du dictionnaire 'full_dict'
for titre in full_dict:
    # Pour chaque titre dans 'full_dict' ajouter une colonne du même nom dans le DF 'daily_df' qui correspondra au retour
    # journalier de ce titre
    daily_df[titre] = full_dict[titre]["daily_return"].loc[1:] # Pour éviter de prendre la première ligne de valeurs 'NaN'
    # Ajouter UNE colonne de date au DF 'daily_df' afin de situer chaque retour journalier dans le temps
    # Utilser la valeur du compteur pour assurer que seulement une colonne de date sera ajoutée
    while i == 0:
        daily_df["Date"] = full_dict[titre]["Date"].loc[1:]
        i = i + 1

# Transformer le type de la colonne "Date" du DF 'daily_df' en datetime64
daily_df["Date"] = pd.to_datetime(daily_df["Date"])
       
# Mettre la date de chaque ligne en index en utilisant .set_index()
daily_df = daily_df.set_index("Date")

# Calculer le rendement annuel individuel de chaque actif
# Pour ce faire, calculer la moyenne de chaque colonne du DF 'daily_df'
# Cependant, comme nous cherchons le rendement annuel, il faudra appliquer la fonction resample('Y') avant ('Y' signifie
# 'Year')
rend_indv = daily_df.resample('Y').last().mean()
# Afficher le rendement annuel individuel de chaque titre
print("Voici les rendements annuels moyens des titres du secteur {0}:".format(secteur_dict[direction]))
rend_indv
```

```python
# Créer la matrice des covariances
# Cette étape consiste à utiliser la même logique que dans l'étape précédente, mais cette fois-ci sur la colonne 'Close'
# des DF qui sont les valeurs du dictionnaire 'full_dict'
close_df = pd.DataFrame()

counter = 0

for asset in full_dict:
    close_df[asset] = full_dict[asset]["Close"].loc[1:] # Pour éviter de prendre la première ligne de valeurs 'NaN'
    while counter == 0:
        close_df["Date"] = full_dict[asset]["Date"].loc[1:]
        counter = counter + 1

# Transformer le type de la colonne "Date" du DF 'close_df' en datetime64
close_df["Date"] = pd.to_datetime(close_df["Date"])
       
# Mettre la date de chaque ligne en index en utilisant .set_index()
close_df = close_df.set_index("Date")

# Créer et afficher la matrice des covariances
# La fonction pct_change() calcule le pourcentage d'augmentation ou de diminution de valeurs consécutives
# On utilise un fonction lambda qui va calculer le logarithme naturel du pourcentage de changement
# La fonction cov() créé la matrice de covariance
matrice_cov = close_df.pct_change().apply(lambda x: np.log(1+x)).cov()
print("Voici la matrice des covariance des titres dans le secteur {0}:".format(secteur_dict[direction]))
matrice_cov
```

```python
# Créer un 'array' vide où on stockera les différents rendements de portefeuilles
port_rend = []
# Créer un 'array' vide où on stockera les différentes volatilités de portefeuilles
port_vol = []
# Créer un portefeuille vide où on stockera les différents poids de titres pour chaque portefeuille
port_poids = []
# Nombre d'actifs dans le portefeuille = Longueur du DF 'allocation' créé selon la direction entrée par l'utilisateur
nb_actifs = len(allocation.columns)
# Nombre de portefeuilles simulés qui seront utilisés pour construire la frontière efficace (10000)
nb_port = 10000

# Pour créer les portefeuilles simulés nous aurons besoin d'une boucle 'for' qui va itérer sur un 'range' entre 0 et 10000
for portefeuille in range(nb_port):
    # Utiliser np.random.random() pour géner une gamme de poids aléatoire pour chaque titre dans le portefeuille
    # entre 0 et 1
    poids = np.random.random(nb_actifs)
    # Puisque la somme des poids du portefeuille doit être égale à 1, il faut diviser chaque poids par la somme des poids
    poids = poids / np.sum(poids)
    # Ajouter chaque poids au 'array' vide 'port_poids'
    port_poids.append(poids)
    # Calculer le rendement de chaque portfolio - C'est le produit des rendements individuels moyens et leurs poids
    returns = np.dot(poids, rend_indv)
    # Ajouter chaque rendement de portfolio au 'array' vide 'prt_rend'
    port_rend.append(returns)
    # Calculer la variance de chaque portefeuille
    variance = matrice_cov.mul(poids, axis = 0).mul(poids, axis = 1).sum().sum()
    # Calculer l'écart-type journalier de chaque portefeuille
    daily_et = np.sqrt(variance)
    # Calucler l'écart-type annuel de chaque portefeuille (donc leur volatilité annuelle)
    # On utilise 250 ici parce qu'il y a environ 250 jours de 'trading' par année
    annual_et = daily_et * np.sqrt(250)
    # Ajouter chaque écart-type annuel au 'array' vide 'port_vol'
    port_vol.append(annual_et)

# Créer un DataFrame ayant le 'Rendement' et la 'Volatilité' comme deux premières colonnes et le poids de chaque titre
# pour un portefeuille donné pour les autres colonnes

# D'abord, il faut mettre les rendements et écart-types créés plus tôt dans un dictionnaire
eff_data = {"Rendement": port_rend, "Volatilité": port_vol}

# Rajouter les colonnes de poids
# Utiliser une boucle 'for' pour itérer sur chaque titre dans le DF 'allocation'. On utilise .columns.tolist() pour mettre
# les noms des titres dans une liste
for index, ticker in enumerate(allocation.columns.tolist()):
    # Ajouter la colonne "ticker poids" (ex: "ADM poids") au dictionnaire 'eff_data'
    # Puisque les noms des colonnes sont dans le même ordre que les poids dans "port_poids", on peut utiliser une liste
    # par compréhension pour prendre le poids de chaque titre
    eff_data[ticker + " poids"] = [weight[index] for weight in port_poids]
    
# Nous pouvons maintenant créer et afficher le DF qui servira à créer la frontière efficiente
portefeuilles = pd.DataFrame(eff_data)
print("Voici le DF contenant les 10000 portefeuilles aléatoires créés pour le secteur {0}:".format(secteur_dict[direction]))
portefeuilles.head()    
  
```

```python
# On créée le dataframe qui regroupe la variable dépendante avec les variables explicatives en faisant correspondre la date
# de leur enregistrement

analysis_df = pd.merge(variable_dependante_df, variables_independantes_df, left_on="Date", right_on="date", how='left')
analysis_df.drop(columns="date", inplace=True)
analysis_df.set_index("Date", inplace=True)


# On cherche a voir si il y a des valeurs nulles
if analysis_df.isnull().any().any():
    analysis_df_na = pd.DataFrame()
    analysis_df_na = analysis_df.copy()
    analysis_df.dropna(inplace=True)
    print("- VALEURS MANQUANTES DETECTEES.\nSUPRESSION DE {} LIGNE(S)".format(len(analysis_df_na) - len(analysis_df)))
```

```python
# Après avoir testé nos données pour les valeurs nulles, nous allons tester la multicollinarité.
# C'est à dire, voir si nos variables explicatives sont corrélé entre elle

# Cette fonction affiche la matrice de correlation pour les variables du dataframe passé en paramettre
# Aussi, une visualisation de la matrice de correlation est affiché
def correlation_matrix_rm(df):

    # calculate the correlation matrix
    corr = df.corr()

    # display the correlation matrix
    display(corr)

    # plot the correlation heatmap
    sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
    
    return None


# On appelle la fonction sur notre tableau d'analyse, avant toutes corrections
correlation_matrix_rm(analysis_df)
```

```python
# Un façon de réduire la collinéarité de notre modèle est de retiré les variables se comportant de manière trop similaire
# pour ce faire, le code ci dessous retire les variables ayant un facteur d'inflation de variance suppérieur au seuil choisie


# On choisit le seuil pour notre VIF
SEUIL_VIF = 10

# On conserve une copie du dataframe, avant toute modification
analysis_df_before = analysis_df.copy()

# La constante X1 généré par le module statmat est nécessaire pour trouver le VIF, aussi calculé par une méthode du module
X1 = sm.tools.add_constant(analysis_df_before)

# On crée une série contenant le VIF pour toute les variables...
series_before = pd.Series([variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])], index=X1.columns)

# ... et on affiche cette série
print('DATA BEFORE')
print('-'*100)
display(series_before)
#correlation_matrix_rm(analysis_df)

# On place les noms des variables explicatives ayant un VIF suppérieur au seuil dans la list "BadVariables"
s = series_before.where(series_before > SEUIL_VIF)
s.dropna(inplace=True)
bad_variables = s.index.tolist()
bad_variables.remove("Close")
bad_variables.remove("const")

# On signale à l'utilisateur si des badVariables ont été trouvé
if len(bad_variables) > 0:
    print("MULTICOLINEARITE DETECTEE DANS LES VARIABLES EXPLICATIVES\nSUPPRESSION DE {} VARIABLES".format(len(bad_variables)))
    for i in bad_variables:
        print("\t{}- supprimé".format(i))

# On retire les variables répertorié comme BadVariables du dataframe d'analyse
analysis_df_after = analysis_df.copy()
for i in bad_variables:
    analysis_df_after.drop(i, axis=1, inplace=True)

print('DATA AFTER')
print('-'*100)
#display(series_after)
#print(analysis_df_after)

correlation_matrix_rm(analysis_df_after)
```

```python
# Sommaire statistique de notre tableau d'analyse
analysis_desc = analysis_df_after.describe()

# standard deviation metric
# analysis_desc.loc['+3_std'] = analysis_desc.loc['mean'] + (analysis_desc.loc['std'] * 3)
# analysis_desc.loc['-3_std'] = analysis_desc.loc['mean'] - (analysis_desc.loc['std'] * 3)

# 
#print(analysis_desc)

# analysis_clean_df est un dataframe dans lequel, pour chaque variable, toute valeurs à plus de 3 écart types de
# la moyenne est considéré comme valeur extrême, et donc est retiré
analysis_clean_df = analysis_df_after[(np.abs(stats.zscore(analysis_df_after)) < 3).all(axis=1)]

# On affiche le nombre d'enregistrement contenant des valeurs extremes
print("VALEURS EXTREMES DETECTEE\nSUPPRESSION DE {} LIGNE(S)"
      .format(len(analysis_df.index.difference(analysis_clean_df.index))))

```

```python
# Pour effectuer la regression multiple, nous avons recours aux modules sklearn

# Pour effectuer l'analyse, nous devons definir nos variables explicative (vecteur X) et une variable réponse (vecteur Y)
# Pour X, nous prenons toutes les variables de notre tableau d'analyse à l'exeption de la variable dépendante
X = analysis_clean_df.drop('Close', axis = 1)

# Pour Y, nous prenons la variable dépendante de notre tableau d'analyse
Y = analysis_clean_df[['Close']]


# TODO: Transformation de Y pour améliorer le fit du modèle

#hetero_fix = analysis_clean_df.copy()
#a, b = stats.boxcox(list(hetero_fix['Close']))
#hetero_fix["Close"] = a
#Y = pd.DataFrame(hetero_fix["Close"])

# on effectue la regression
# TODO: check la docu train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=1)


regression_model = LinearRegression()

# pass through the X_train & y_train data set
regression_model.fit(X_train, y_train)
```

```python
# On stock la valeur du x-intercept ainsi que les coefficients des variables explicatives
intercept = regression_model.intercept_[0]
coefficent = regression_model.coef_[0][0]

# On affiche l'interception
print("Intercept {:.4}".format(intercept))
print('-'*100)

# on affiche le coefficient de chaque variable indépendante
for coef in zip(X.columns, regression_model.coef_[0]):
    print("The Coefficient for {} is {:.2}".format(coef[0],coef[1]))
    
    
# Le module nous permet de faire des prediction
# TODO: check docu
y_predict = regression_model.predict(X_test)
y_predict[:5]
```

```python
# Afin d'évaluer notre modèle nous allons utilisé le module statmat, qui facilite de nombreux test statistique

# On doit modifier notre matrice X en X2 afin de l'utiliser avec statmat
X2 = sm.add_constant(X)

# création du modele OLS
model = sm.OLS(Y, X2)

# fit the data
est = model.fit()
```

```python
# On teste l'heteroscédacité des résidus de notre généré par notre modèle. Pour ce faire nous faisons un test breush-pagan

# Breusch-Pagan test
_, pval, __, f_pval = diag.het_breuschpagan(est.resid, est.model.exog)
print(pval, f_pval)
print('-'*100)

# results of the test
if pval > 0.05:
    print("Breusch-Pagan's Test")
    print("p-value: {:.4}".format(pval))
    print("HETEROSCEDACITE NON DETECTEE")

else:
    print("Breusch-Pagan's Test")
    print("p-value: {:.4}".format(pval))
    print("HETEROSCEDACITE DETECTEE")
```

```python
# test d'autocorrelation
from statsmodels.stats.stattools import durbin_watson

# calculate the lag, optional
lag = min(10, (len(X)//5))
print('The number of lags will be {}'.format(lag))
print('-'*100)

# run the Ljung-Box test for no autocorrelation of residuals
# test_results = diag.acorr_breusch_godfrey(est, nlags = lag, store = True)
test_results = diag.acorr_ljungbox(est.resid, lags = lag)

# grab the p-values and the test statistics
ibvalue, p_val = test_results

# print the results of the test
if min(p_val) > 0.05:
    print("lowest p-value: {:.4}".format(min(p_val)))
    print("AUTOCORRELATION NON DETECTEE")
    print('-'*100)
else:
    print("lowest p-value: {:.4}".format(min(p_val)))
    print("AUTOCORRELATION DETECTEE")
    print('-'*100)

# plot autocorrelation
sm.graphics.tsa.plot_acf(est.resid)
plt.show()
```

```python
import pylab

# Test pour la normalité des résidus du model
sm.qqplot(est.resid, line='s')
pylab.show()

# la moyenne des résidus devrait s'approcher de 0
mean_residuals = sum(est.resid)/ len(est.resid)
print("Moyenne des résidus: {:.4}".format(mean_residuals))
```
