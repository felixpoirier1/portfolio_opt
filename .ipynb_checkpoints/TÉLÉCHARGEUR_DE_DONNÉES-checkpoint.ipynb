```python
import pandas as pd 
import datetime as dt
import time
# pip install -i https://pypi.anaconda.org/ranaroussi/simple yfinance
# pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
import os
```

```python
df = pd.read_csv('secteurs.csv', sep=";")
secteurs = df.columns
```

```python
# Fonction qui créer un dataframe et qui l'enregistre en csv
def save_to_csv_from_yahoo(folder, ticker):
    stock = yf.Ticker(ticker)
    print(stock)
    try:
        # données sur 5 ans
        df = stock.history(period="5y")
        
        
        # Wait 1 seconde (pour éviter de bombarder les serveurs de Yahoo)
        #time.sleep(1)

        # Nom du fichiers
        the_file = folder + ticker.replace(".", "_") + '.csv'
            
        
        df.to_csv(the_file)
        
        # voir si dataframe est vide
        try:
            df.iloc[0]
        
        # supprimer fichier sinon
        except IndexError:
            print('\tFile is empty')
            os.remove(the_file)
            return 0
        # supprimer si ne commence pas il y a 5 ans            
        if df.index[0] != pd.Timestamp('2017-04-03 00:00:00'):
            print('\tData does not go back far enough')
            os.remove(the_file)
            return 0
        
        # supprimer si données pas récentes      
        if df.index[-1] != pd.Timestamp('2022-04-01 00:00:00'):
            print('\tData does not go far enough')
            os.remove(the_file)
            return 0
        
        print(the_file)
        
    except Exception as ex:
        print("\tCouldn't Get Data for :", ticker)
        
```

```python
# pour télécharger les stocks
for secteur in secteurs:
    lieu = f'STOCKS/{secteur.upper()}/'
    if os.path.exists(lieu):
        pass
    else:
        os.mkdir(lieu)
    
    for stock in df.loc[:][secteur]:
        if type(stock) != float:
            save_to_csv_from_yahoo(lieu, stock)
```

```python
df = pd.read_csv('index.csv', sep=";")
indexs = df.columns

# pour télécharger les index
for index in indexs:
    lieu = f'STOCKS/{index.upper()}/'
    
    for index in df.loc[:][index]:
        if type(index) != float:
            save_to_csv_from_yahoo(lieu, index)
            
        fichier = lieu + index.replace(".", "_") + '.csv'
```
